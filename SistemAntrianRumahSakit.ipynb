```python
antrian_umum = 0
antrian_gigi = 0
antrian_kandungan = 0
antrian_psikolog = 0
antrian_jantung = 0

while True:
    print ("-------------------------------------------------")
    print("Silahkan Registrasi Pasien Untuk Mengambil Antrian")
    print("--------------------------------------------------")
    print("1. Dokter Umum")
    print("2. Dokter Gigi")
    print("3. Dokter Kandungan")
    print("4. Dokter Psikolog")
    print("5. Dokter Jantung")
    print("--------------------------------------------------")
    pilihan1 = input(str("Masukkan Pilihan Anda 1/2/3/4/5 : "))

    if pilihan1 == str(1):
      print("\nAntrian Pasien Dokter Umum")
      print("--------------------------")
      nama_pasien_umum = input(str("Nama Pasien : "))
      umur_pasien_umum = input(str("Umur Pasien : "))
      antrian_umum += 1
      waktuantrianumum = antrian_umum * 10
      print("--------------------------")
      print("Nomor Antrian Anda : ", antrian_umum)
      print("Waktu Tunngu: ", waktuantrianumum , "menit. Silahkan Menunggu TerimaKasih.")

    elif pilihan1 == str(2):
      print("\nAntrian Pasien Dokter Gigi")
      print("--------------------------")
      nama_pasien_gigi = input(str("Nama Pasien : "))
      umur_pasien_gigi = input(str("Umur Pasien : "))
      antrian_gigi += 1
      waktuantriangigi = antrian_gigi * 10
      print("--------------------------")
      print("Nomor Antrian Anda : ", antrian_gigi)
      print("Waktu Tunngu: ", waktuantriangigi , "menit. Silahkan Menunggu TerimaKasih.")

    elif pilihan1 == str(3):
      print("\nAntrian Pasien Dokter Kandungan")
      print("--------------------------")
      nama_pasien_kandungan = input(str("Nama Pasien : "))
      umur_pasien_kandungan = input(str("Umur Pasien : "))
      antrian_kandungan += 1
      waktuantriankandungan = antrian_kandungan * 10
      print("--------------------------")
      print("Nomor Antrian Anda : ", antrian_kandungan)
      print("Waktu Tunngu: ", waktuantriankandungan , "menit. Silahkan Menunggu TerimaKasih.")

    elif pilihan1 == str(4):
      print("\nAntrian Pasien Dokter Psikolog")
      print("--------------------------")
      nama_pasien_psikolog = input(str("Nama Pasien : "))
      umur_pasien_psikolog = input(str("Umur Pasien : "))
      antrian_psikolog += 1
      waktuantrianpsikolog = antrian_psikolog * 10
      print("--------------------------")
      print("Nomor Antrian Anda : ", antrian_psikolog)
      print("Waktu Tunngu: ", waktuantrianpsikolog , "menit. Silahkan Menunggu TerimaKasih.")

    elif pilihan1 == str(5):
      print("\nAntrian Pasien Dokter Jantung")
      print("--------------------------")
      nama_pasien_jantung = input(str("Nama Pasien : "))
      umur_pasien_jantung = input(str("Umur Pasien : "))
      antrian_jantung += 1
      waktuantrianjantung = antrian_jantung * 10
      print("--------------------------")
      print("Nomor Antrian Anda : ", antrian_jantung)
      print("Waktu Tunngu: ", waktuantrianjantung , "menit. Silahkan Menunggu TerimaKasih.")

    kel_ap = input("apakah anda ingin keluar? y/n :")
    if kel_ap.lower()== "y":
        print("TerimaKasih. Anda telah keluar aplikasi")
        break
    elif kel_ap.lower()== "n":
      continue
    else :
      print("Silahkan Masukkan Pilihan Anda")


```
